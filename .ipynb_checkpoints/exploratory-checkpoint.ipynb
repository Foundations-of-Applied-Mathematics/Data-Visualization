# Exploratory Visualizations

```python
import pandas as pd
import numpy as np
import pydataset
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pydataset.data('Salaries')
df.head()
```

```python
df.plot(kind='scatter',x='yrs.since.phd',y='salary',color='r',alpha=.5,linewidth=0)
```

```python
df['yrs.since.phd'].plot(kind='hist',bins=30,color='r',alpha=.5,linewidth=0)
```

```python
df['salary'].plot(kind='hist',bins=50,color='r',alpha=.5,linewidth=0)
```

```python
df2 = pydataset.data('diamonds')
df2.head()
```

```python
df2.plot(kind='scatter',x='price',y='carat',linewidth=0)
```

```python
df2.plot(kind='scatter',x='price',y='carat',linewidth=0,alpha=.3)
```

```python
df2.plot(kind='scatter',x='price',y='carat',linewidth=0,alpha=.3,figsize=(8,8))
```

```python
df2.plot(kind='scatter',x='price',y='carat',linewidth=0,alpha=.3,figsize=(8,8))
plt.xlim(0,20000)
plt.ylim(0,2)
plt.show()
```
