```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
## matplotlib is version 1.5
```

```python
data = plt.imread('calf.jpg')
```

```python
plt.imshow(data)
#plt.colorbar()
plt.show()
plt.imshow(data[:,:,0],cmap='jet')
plt.colorbar()
plt.show()
plt.imshow(data[:,:,0],cmap='viridis')
plt.colorbar()
plt.show()
plt.imshow(data[:,:,0],cmap='magma')
plt.colorbar()
plt.show()
```

```python
data2 = plt.imread('calfjet.png')
```

```python
# this is what jet looks like in black and white
plt.imshow(data2[:,:,0],cmap='Greys_r')
plt.show()
```

```python
# compared to the image straight to black and white
plt.imshow(data[:,:,0],cmap='Greys_r')
plt.colorbar()
plt.show()
# you can see more contrast in the background grass in the jet grayscale plot...
# but that's about all that's good about it
```

```python
# use to show differences
plt.imshow(data[:,:,0],cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
# a very good sequential plot
plt.imshow(data[:,:,0],cmap='afmhot')
plt.colorbar()
plt.show()
```
