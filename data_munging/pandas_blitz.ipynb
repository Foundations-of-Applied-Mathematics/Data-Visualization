```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Pandas uses numpy to handle data in a really nice way. It's pretty easy to plot stuff because pandas also uses matplotlib. Here's an example of loading a data set using pandas. I got this data from data.gov under the health care section. It's called hospital charge data on the site.

```python
# df stands for dataframe, which is a pandas object designed to work with data
# you don't need to call it df
# I'm creative with visualizations not variable names
df = pd.read_csv("hospital_charge.csv")
```

```python
# you can take a peek at the first 5 rows of your data like this
df.head()
```

```python
# or the last 5 rows like this
df.tail()
```

```python
# or a random sample of your data like this
df.sample(5)
```

```python
# if your data set already has a column, you can do this
df2 = pd.read_csv("hospital_charge.csv",index_col=0)
df2.head()
# this creates a new dataframe object with column number 0 as the index
# this can be whatever number you want to set as your index
```

```python
# you can specify columns names like this
L = ['there','goes','john','jacob','jingle','heimer','schmidt','bingo','was','his','name','oh']
df3 = pd.read_csv("hospital_charge.csv",names=L)
df3.head()
```

```python
# if you cleaned your data in numpy first, you can stick it in a dataframe
random_data = np.random.randn(100)
df4 = pd.DataFrame(random_data,columns=['first'])
df4.head()
```

```python
# if you want to add a new row, just do this
df4['second'] = np.random.randint(100)
df4.head()
```

```python
# you can also manipulate columns
df4['third'] = df4['first']*df4['second']
df4.head()
```

```python
# plotting looks like this
df4.plot()
```

```python
# pretty ugly but we'll fix that later
# another example
df4.plot(kind='scatter',x='first',y='third')
```
